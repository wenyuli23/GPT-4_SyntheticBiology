```python
def r_squared(y_true, y_pred):
    """Custom R2 scorer. Takes ndarray as input"""
    tss = np.sum((y_true - np.mean(y_true)) ** 2)
    rss = np.sum((y_true - y_pred) ** 2)
    
    r2 = 1 - (rss/tss)
    return r2

def plot_od(y_train, y_train_pred, y_test, y_test_pred, alg, scoring):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)

    ax1.scatter(y_train, y_train_pred, s=10, c='b', marker="s", label='train')
    ax1.scatter(y_test, y_test_pred, s=10, c='r', marker="o", label='test')

    plt.legend(loc='upper right');
    plt.xlabel('observed OD')
    plt.ylabel('prediction OD')
    plt.savefig(f"./figures/{alg}_od_{scoring}.png")
    
def plot_gridsearchcv_results(estimator, alg):
    cv_results = pd.DataFrame(estimator.cv_results_)
    print(cv_results.columns)
    cv_results['params'] = cv_results['params'].astype(str)

    # Melt the df into a long format
    melted_df = cv_results.melt(
        id_vars=['params', 'rank_test_rmse', 'rank_test_mae', 'rank_test_r2'],
        value_vars=[
            col for col in cv_results.columns if col.startswith('split') and col.endswith(('rmse', 'mae', 'r2'))
        ],
        var_name='Metric_Fold', value_name='Value'
    )

    metric_fold_split = melted_df['Metric_Fold'].str.extract(r'(split\d+_test_)(.*)')
    melted_df['Fold'] = metric_fold_split[0].str.extract('(\d+)')
    melted_df['Metric'] = metric_fold_split[1]
    melted_df.loc[melted_df['Metric'].isin(['rmse', 'mae']), 'Value'] *= -1
    print(melted_df.info())

    # Plot
    g = sns.FacetGrid(melted_df, col='params', col_wrap=2, sharey=False, height=4, aspect=1.5)
    g.map_dataframe(sns.barplot, x='Fold', y='Value', hue='Metric', palette='deep')
    g.add_legend(title='Metric')
    g.set_axis_labels('Fold', 'Value')
    g.set_titles(col_template="{col_name}")

    plt.suptitle(f'Cross-validation results for {alg}', y=1.05)
    plt.savefig("cv_result.png")
    plt.show()

# assume optimizing r2
def plot_prediction(product_class, X_test_plt, y_test, y_pred, alg):
    class_names = {
        1: "Organic acid",
        2: "Lipid",
        3: "Small terpene (MW<300 g/mol)",
        4: "Large terpene (MW>300 g/mol)",
        5: "Flavonoid",
        6: "Fatty acid-derived",
        7: "Sugar alcohol",
        8: "Glycan",
        9: "Polyketide",
        10: "Biomass",
    }

    palette = [
        '#377eb8', # Blue
        '#4daf4a', # Green
        '#e41a1c', # Red
        '#ff7f00', # Orange
        '#984ea3', # Purple
        '#a65628', # Brown
        '#f781bf', # Pink
        '#a6cee3', # Light Blue
        '#46f0f0', # Cyan
        '#999999'  # Gray
    ]

    fig, ax = plt.subplots(figsize=(6, 6))

    for idx, (product_class, class_name) in enumerate(class_names.items()):
        # Filter test data and predictions for the current product class
        mask = X_test_plt['plot'] == product_class
        filtered_test_data = y_test[mask]
        filtered_predictions = y_pred[mask]
        ax.scatter(filtered_test_data, filtered_predictions, color=palette[idx], label=f"{class_name}", s=10)

    font_size = 12
    font_prop = font_manager.FontProperties(family='Arial', size=font_size)

    ax.set_xlabel('Test data titer (g/L)', fontproperties=font_prop)
    ax.set_ylabel('ML prediction titer (g/L)', fontproperties=font_prop)
    ax.set_title(f"{alg} test prediction", fontproperties=font_prop)
    ax.set_xscale('log')
    ax.set_yscale('log')

    ax.tick_params(axis='both', which='major', labelsize=font_size)
    for label in ax.get_xticklabels():
        label.set_fontproperties(font_prop)
    for label in ax.get_yticklabels():
        label.set_fontproperties(font_prop)

    # Add a 1-to-1 line
    max_value = max(y_test.max(), y_pred.max())*3
    ax.plot([0, max_value], [0, max_value], color='black', linestyle='--', label='1-to-1 Line')

    ax.set_xlim(1e-5, max_value)
    ax.set_ylim(1e-5, max_value)

    # Move the legend outside the plot
    font = font_manager.FontProperties(family='Arial', size=12)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), prop=font)

    plt.savefig(f'../figures/{alg} scatter', dpi=300, bbox_inches='tight')

def run_model(regressor, param_grid, train_x, train_y, test_x, test_y, scoring, figname):
    print(f"\n scoring: {scoring}\n")
    # 5-fold cv
    grid_search = GridSearchCV(estimator = regressor, param_grid = param_grid, n_jobs = -1, verbose = 3, refit=True, scoring=scoring, error_score='raise')
    grid_search.fit(train_x, train_y)
    
    # reverse y
    #y_train_hat = np.exp(grid_search.predict(train_x.values))
    #y_test_hat = np.exp(grid_search.predict(test_x.values))
    y_train_hat = grid_search.predict(train_x)
    y_test_hat = grid_search.predict(test_x)
    print("Average true labels: ", y_test_hat.mean())
    print("Average predicted values", test_y.mean())
    
    print("Best score in grid search: ", grid_search.best_score_)
    print("Chosen params based on hold-out: ", grid_search.best_params_)
    
    print("----------TRAIN----------")
    rmse_train = mean_squared_error(train_y, y_train_hat, squared=False)
    mae_train = mean_absolute_error(train_y, y_train_hat)
    r2_train = r_squared(train_y, y_train_hat)
    mape_train = mean_absolute_percentage_error(train_y, y_train_hat)
    print("RMSE =", rmse_train)
    print("MAE =", mae_train)
    print("r2 =", r2_train)
    print("MAPE =", mape_train)
    train_scores = [rmse_train, mae_train, r2_train, mape_train]
    
    print("----------TEST----------")
    rmse_test = mean_squared_error(test_y, y_test_hat, squared=False)
    mae_test = mean_absolute_error(test_y, y_test_hat)
    r2_test = r_squared(test_y, y_test_hat)
    mape_test = mean_absolute_percentage_error(test_y, y_test_hat)
    print("RMSE =", rmse_test)
    print("MAE =", mae_test)
    print("r2 =", r2_test)
    print("MAPE =", mape_test)
    test_scores = [rmse_test, mae_test, r2_test, mape_test]
    return train_scores, test_scores, grid_search, y_test_hat
```
