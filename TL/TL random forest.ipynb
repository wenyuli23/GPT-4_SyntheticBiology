```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import matplotlib.font_manager as font_manager
from matplotlib.font_manager import FontProperties


%config InlineBackend.figure_format = 'retina'
%matplotlib inline

# Read excel file
df = pd.read_excel('../data/TL training data YL.xlsx', header=1)
dfRT = pd.read_excel('../data/TL training data RT.xlsx', header=1)
dfpred1 = pd.read_excel('../data/TL predict gene number 0.xlsx', header=1)
dfpred2 = pd.read_excel('../data/TL predict gene number 2.xlsx', header=1)
dfpred3 = pd.read_excel('../data/TL predict gene number 4.xlsx', header=1)
dfpred4 = pd.read_excel('../data/TL predict gene number 6.xlsx', header=1)
# The feature names are in the second row (already done in the previous step)


# Set the target column
target = 'Product titer (g/L)'

# Count the semicolons in 'genes_modified_updated'
df['Genes modified'] = df['Genes modified'].fillna('').apply(lambda x: x.count(';'))
dfRT['Genes modified'] = dfRT['Genes modified'].fillna('').apply(lambda x: x.count(';'))
dfpred1['Genes modified'] = dfpred1['Genes modified'].fillna('').apply(lambda x: x.count(';'))
dfpred2['Genes modified'] = dfpred2['Genes modified'].fillna('').apply(lambda x: x.count(';'))
dfpred3['Genes modified'] = dfpred3['Genes modified'].fillna('').apply(lambda x: x.count(';'))
dfpred4['Genes modified'] = dfpred4['Genes modified'].fillna('').apply(lambda x: x.count(';'))

# Count the ones in 'gene_overexpression' and 'heterologous_gene'
df['Genes overexpressed'] = df['Genes overexpressed'].fillna('').apply(lambda x: str(x).count('1'))
df['Heterologous genes'] = df['Heterologous genes'].fillna('').apply(lambda x: str(x).count('1'))
df['Genes deleted'] = df['Genes deleted'].fillna('').apply(lambda x: str(x).count('1'))

dfRT['Genes overexpressed'] = dfRT['Genes overexpressed'].fillna('').apply(lambda x: str(x).count('1'))
dfRT['Heterologous genes'] = dfRT['Heterologous genes'].fillna('').apply(lambda x: str(x).count('1'))
dfRT['Genes deleted'] = dfRT['Genes deleted'].fillna('').apply(lambda x: str(x).count('1'))

dfpred1['Genes overexpressed'] = dfpred1['Genes overexpressed'].fillna('').apply(lambda x: str(x).count('1'))
dfpred1['Heterologous genes'] = dfpred1['Heterologous genes'].fillna('').apply(lambda x: str(x).count('1'))
dfpred1['Genes deleted'] = dfpred1['Genes deleted'].fillna('').apply(lambda x: str(x).count('1'))

dfpred2['Genes overexpressed'] = dfpred2['Genes overexpressed'].fillna('').apply(lambda x: str(x).count('1'))
dfpred2['Heterologous genes'] = dfpred2['Heterologous genes'].fillna('').apply(lambda x: str(x).count('1'))
dfpred2['Genes deleted'] = dfpred2['Genes deleted'].fillna('').apply(lambda x: str(x).count('1'))

dfpred3['Genes overexpressed'] = dfpred3['Genes overexpressed'].fillna('').apply(lambda x: str(x).count('1'))
dfpred3['Heterologous genes'] = dfpred3['Heterologous genes'].fillna('').apply(lambda x: str(x).count('1'))
dfpred3['Genes deleted'] = dfpred3['Genes deleted'].fillna('').apply(lambda x: str(x).count('1'))

dfpred4['Genes overexpressed'] = dfpred4['Genes overexpressed'].fillna('').apply(lambda x: str(x).count('1'))
dfpred4['Heterologous genes'] = dfpred4['Heterologous genes'].fillna('').apply(lambda x: str(x).count('1'))
dfpred4['Genes deleted'] = dfpred4['Genes deleted'].fillna('').apply(lambda x: str(x).count('1'))

# Remove 'dataID' and 'paper_number' columns
df = df.drop(['dataID', 'paper_number','product_name'], axis=1)
dfRT = dfRT.drop(['dataID', 'paper_number','product_name'], axis=1)
dfpred1 = dfpred1.drop(['dataID', 'paper_number','product_name'], axis=1)
dfpred2 = dfpred2.drop(['dataID', 'paper_number','product_name'], axis=1)
dfpred3 = dfpred3.drop(['dataID', 'paper_number','product_name'], axis=1)
dfpred4 = dfpred4.drop(['dataID', 'paper_number','product_name'], axis=1)

# Impute all NaN data as zero
imputer = SimpleImputer(strategy='constant', fill_value=0)
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
dfRT_imputed = pd.DataFrame(imputer.fit_transform(dfRT), columns=df.columns)  # Impute dfRT with the same imputer
dfpred1_imputed = pd.DataFrame(imputer.fit_transform(dfpred1), columns=dfpred1.columns)
dfpred2_imputed = pd.DataFrame(imputer.fit_transform(dfpred2), columns=dfpred2.columns)
dfpred3_imputed = pd.DataFrame(imputer.fit_transform(dfpred3), columns=dfpred3.columns)
dfpred4_imputed = pd.DataFrame(imputer.fit_transform(dfpred4), columns=dfpred4.columns)

# Define X, y, and product_class for df
X = df_imputed.drop(target, axis=1)
y = df_imputed[target]
product_class = df_imputed['Product class']

# Define X_RT, y_RT, and product_class_RT for dfRT
X_RT = dfRT_imputed.drop(target, axis=1)
y_RT = dfRT_imputed[target]
product_class_RT = dfRT_imputed['Product class']

# Drop the target column from the prediction data
dfpred1_imputed = dfpred1_imputed.drop(target, axis=1)
dfpred2_imputed = dfpred2_imputed.drop(target, axis=1)
dfpred3_imputed = dfpred3_imputed.drop(target, axis=1)
dfpred4_imputed = dfpred4_imputed.drop(target, axis=1)

# Initialize predictions lists
preds_dfpred1 = []
preds_dfpred2 = []
preds_dfpred3 = []
preds_dfpred4 = []

# Stratified data split
n_splits = 1000
split = StratifiedShuffleSplit(n_splits=n_splits, test_size=0.2, random_state=666*6)

# Define the weight for dfRT data
dfRT_weight = 3  # change this to the desired weight

# In the loop of StratifiedShuffleSplit
for train_index, test_index in split.split(X, product_class):
    # Split the data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Duplicate dfRT data with the desired weight
    dfRT_X_weighted = pd.concat([dfRT_imputed.drop(target, axis=1)] * dfRT_weight, ignore_index=True)
    dfRT_y_weighted = pd.concat([dfRT_imputed[target]] * dfRT_weight, ignore_index=True)

    # Combine weighted dfRT with X_train
    X_train = pd.concat([X_train, dfRT_X_weighted], axis=0)
    y_train = pd.concat([y_train, dfRT_y_weighted], axis=0)

    # Train random forest regression
    rf = RandomForestRegressor()
    rf.fit(X_train, y_train)

    # Predict on dfpred1 and dfpred2
    preds_dfpred1.append(rf.predict(dfpred1_imputed))
    preds_dfpred2.append(rf.predict(dfpred2_imputed))
    preds_dfpred3.append(rf.predict(dfpred3_imputed))
    preds_dfpred4.append(rf.predict(dfpred4_imputed))
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Set default font to Arial
mpl.rcParams['font.sans-serif'] = 'Arial'
mpl.rcParams['font.family'] = 'sans-serif'

def plot_frequency_distribution(predictions, title, color, ax):
    concatenated_preds = np.concatenate(predictions)
    ax.hist(concatenated_preds, bins=100, density=True, alpha=0.6, color=color)
    ax.set_xlabel('Predicted titer (g/L)')
    ax.set_ylabel('Probability density')
    ax.set_title(title)

    # Calculate the average and plot a vertical line
    avg = np.mean(concatenated_preds)
    ax.axvline(avg, color='k', linestyle='dashed', linewidth=1)
    ax.text(avg, ax.get_ylim()[1] * 0.9, f'Average: {avg:.4f}', rotation=0, color='k')

# Create a shared figure with three subplots
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(6, 7), sharex=True, sharey=False ,dpi=300)

# Plot the frequency distributions
plot_frequency_distribution(preds_dfpred1, '', '#e7cfd4', ax1)
plot_frequency_distribution(preds_dfpred2, '', '#cf9198', ax2)
plot_frequency_distribution(preds_dfpred3, '', '#d89c7c', ax3)
plot_frequency_distribution(preds_dfpred4, '', '#aa3e53', ax4)

# Set x-axis tick label visibility
ax1.xaxis.set_tick_params(which='both', labelbottom=True)
ax2.xaxis.set_tick_params(which='both', labelbottom=True)
ax3.xaxis.set_tick_params(which='both', labelbottom=True)
ax4.xaxis.set_tick_params(which='both', labelbottom=True)

# Adjust layout and display the figure
plt.tight_layout()
plt.show()
```
